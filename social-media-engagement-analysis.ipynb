```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("social_media_engagement_5000.csv")
df
```

```python
df['posted_at']=pd.to_datetime(df['posted_at'],format='%d-%m-%Y')
```

```python
df["age"] = df["age"].astype("Int64")
```

```python
df[['gender','country','post_type','post_category','device_type','sentiment']]=df[['gender','country','post_type','post_category','device_type','sentiment']].astype('category')
```

```python
df[["likes", "comments", "shares"]] = df[["likes", "comments", "shares"]].astype("Int64")
```

```python
df.isnull().sum()
```

```python
df.age.fillna(df['age'].median(),inplace=True)
```

```python
df['gender'].fillna(method='ffill',inplace = True)
```

```python


df['gender'].isnull().sum()
```

```python
df['likes'] = df['likes'].fillna(round(df['likes'].median())).astype('Int64')
```

```python
df['likes'].isnull().sum()
```

```python
df['comments'].fillna(df['comments'].mode()[0], inplace=True)
```

```python


df['comments'].isnull().sum()
```

```python
df['shares'].fillna(method='bfill',inplace=True)
```

```python
df['shares'].isnull().sum()
```

```python
df['sentiment'].fillna(df['sentiment'].mode()[0],inplace=True)
```

```python
from matplotlib import pyplot as plt
top=df.head(50).reset_index()
plt.scatter(x=top['likes'],y=top['impression_count'],s=top['engagement_rate']*100,c=top['engagement_rate'])
plt.xlabel('likes')
plt.ylabel('impression_count')
plt.title("likes by impression count")
plt.show()
```

```python
daily_engagement = df.groupby(df["posted_at"].dt.day_name())["engagement_rate"].sum()
weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

# Reindex to enforce order
daily_engagement = daily_engagement.reindex(weekday_order)


plt.figure(figsize=(8, 6))
plt.plot(daily_engagement.index, daily_engagement.values,linestyle='-',color="darkgreen",marker='o')

plt.title("Daily engagement")
plt.xlabel("Day")
plt.ylabel("Total engagement")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
